Âà©Áî®HCÁæ§‰ΩìÈ¶ñÊ¨°Ê†∏Á£ÅÊï∞ÊçÆÂª∫Á´ãwscores

```python
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from pathlib import Path

# ===  Êï∞ÊçÆÊ∏ÖÊ¥ó ===
df['sex'] = df['sex'].astype(str).str.strip().replace({'F':0,'M':1,'Â•≥':0,'Áî∑':1})
df['sex'] = pd.to_numeric(df['sex'], errors='coerce').fillna(0)
df['ICV'] = pd.to_numeric(df['ICV'], errors='coerce')
df['age'] = pd.to_numeric(df['age'], errors='coerce')

# ===  ÊèêÂèñ HC Baseline Êï∞ÊçÆ ===
df_bl = df.loc[df['level_type'] == 'patient-level'].copy()
hc_bl = df_bl[df_bl['group'] == 'f'].copy()

# ===  ÊèêÂèñËÑëÂå∫Âàó ===
print(f"‚úÖ Ê£ÄÊµãÂà∞ {len(sumv_cols)} ‰∏™ËÑëÂå∫Âàó")

# === 6. Âü∫‰∫éHCÁªÑÂª∫Á´ãÂõûÂΩíÊ®°ÂûãÂπ∂ËÆ°ÁÆówÂàÜ ===
w_df = df.copy()

for col in sumv_cols:
    try:
        # Âú®HC Baseline‰∏≠ÊãüÂêàÊ®°Âûã
        model = smf.ols(f"{col} ~ age + sex + ICV", data=hc_bl).fit()
        
        # ËÆ°ÁÆóÂú®ÂÖ®‰ΩìÊ†∑Êú¨‰∏äÁöÑÈ¢ÑÊµãÂÄº
        pred = model.predict(w_df[['age', 'sex', 'ICV']])
        
        # ËÆ°ÁÆówÂàÜ
        resid_std = np.std(model.resid, ddof=1)
        w_col = f"{col}_w"
        w_df[w_col] = (w_df[col] - pred) / resid_std
        
    except Exception as e:
        print(f"‚ö†Ô∏è {col} ÊãüÂêàÂ§±Ë¥•: {e}")
```
